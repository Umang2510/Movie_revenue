```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
3.29e7 #3.29 * 10^7
```

```python
X = DataFrame(data, columns = ['production_budget_usd']) #Feature, Independent Variable
y = DataFrame(data, columns = ['worldwide_gross_usd']) #Target, Dependent Variable
```

```python
plt.figure(figsize=(10,6))
matplotlib.pyplot.scatter(X,y)
plt.title('Film Cost VS Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.show()
```

In above graph you can see some space on X and Y axis it's for a negative values but we will remove this space.

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y, alpha=0.3)
plt.title('Film Cost VS Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X, y)
```

Slope Coefficient:

```python
regression.coef_ #theta_1
```

*   Positive relation between production budget X and Revenue y because coef is positive value.
*   Each dollar we spend on producing a movie we get about 3.11 dollar in revenue





```python
regression.intercept_
```

A movie with 0 dollar budget will actually loose 7 million dollar.

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y, alpha=0.3)
plt.plot(X, regression.predict(X), color='red', linewidth=4)
plt.title('Film Cost VS Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```
